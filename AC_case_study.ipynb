## Step - 2

```python
# Print the statistical and general info about the data

print(df.head())
print(df.info())
print(df.describe())
```

```python
# Print the statistical and general info about the data

print(temp.head())
print(temp.info())
print(temp.describe())
```

```python
# Converting to datetime format

df['Date']= pd.to_datetime(df['0'])
print(df['Date'].head())
```

```python
# Setting the index

df.set_index(df['Date'],inplace=True)
print(df.head())
```

```python
# Drop the column

df.drop(['0','Date'],axis=1,inplace=True)
df.dropna(axis=0,how='all',inplace=True)
df.head()
```

```python
# Print the info of the data

print(df.shape)
print(df.info())
print(df.describe())
```

```python
# Checking for null values

df.isnull().sum()
```

```python
# Using backward filling for the null values

df = df.fillna(method='bfill')
print(df.isnull().sum())
```

## Step - 3

```python
# Creating a new column for the total sum of the power consumed in a minute

df['Total']= df.sum(axis=1)
print(df['Total'].head())
```

```python
# Printing the data

df.head()
```

```python
# Statistical descrpition of the data

df.describe()
```

```python
# To print the temp data

pds=temp.shape[0]
temp['Date']=pd.date_range('2019-08-01',periods=pds,freq='H')
print(temp.head(),temp.tail())
```

```python
# Making a copy of the data

df_min = df.copy()
```

```python
# Resampling the data & then checking for null values

df_h = df.resample('H').sum()
print(df_h.head())
print(df_h.isnull().sum())
```

```python
# Plotting power Vs Time graph

df_min['Total'].plot.line(title= 'Total Power VS Time (in mins)')
```

```python
# Plotting power Vs Time(hourly) graph

df_h['Total'].plot.line(title='Hourly Total Power VS Time')
```

```python
# Plotting power Vs Time(days) graph

df_d = df.resample('D').sum()
df_d['Total'].plot.line(title='Daily Total Power VS Time (in days)')
```

```python
# Finding the max & min consumption 

pw=df_min.iloc[:,0:-1].sum(axis=0)
print(pw)
print(pw.idxmin(),pw.idxmax())
```
